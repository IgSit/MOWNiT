```julia
using DifferentialEquations
using Interact, Plots
```

```julia
function f(du, u, p, t)  # du = [dx, dy], u = [x, y]
    du[1] = - a * u[1] + c * d * u[1] * u[2]
    du[2] = b * u[2] - d * u[1] * u[2]
end
```

```julia
a, b, c, d = 10, 3, 0.2, 0.1
u0 = [10 20]
tspan = (0.0,10)
prob = ODEProblem(f,u0,tspan)
sol = solve(prob, RK4(), reltol=1e-8, abstol=1e-8)

plot(sol,title="Population of predators and prays over time",
     xaxis="day",yaxis="population size",label=["Predrators" "Prays"])
```
